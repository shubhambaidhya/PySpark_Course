```python
# Create DataFrame

emp = spark.createDataFrame(data= emp_data, schema=emp_schema)
```

```python
# Show dataframe (Action)

emp.show()
```

```python
# Print Schema 
emp.printSchema()
```

```python
# Casting Column
# select employee_id, name, age, cast(salary as double) as salary from emp

emp_casted = emp.selectExpr("employee_id", "name", "age", "cast(salary as double) as salary")

emp_casted.printSchema()
```

```python
# Casting Column using col function
# select employee_id, name, age, cast(salary as double) as salary from emp

from pyspark.sql.functions import col

emp_casted = emp.select("employee_id", "name", "age", col("salary").cast("double").alias("salary"))

emp_casted.printSchema()
```

```python
# show dataframe
emp_casted.show()
```

```python
# Adding Columns
# Select employee_id, name, age, salary, (salary * 0.2) as tax from emp_casted
# to add a new column or to overwrite a new column we can use withColumn 

emp_taxed = emp_casted.withColumn("tax", col("salary")*0.2)
```

```python
emp_taxed.show()
```

```python
# Literals
# select employee_id, name, age, salary, tax, 1 as columnOne, 'two' as columnTwo from emp_taxed

from pyspark.sql.functions import lit

emp_new_cols = emp_taxed.withColumn("columnOne", lit(1)).withColumn("columnTwo", lit(2))
```

```python
emp_new_cols.show()
```

```python
# Renaming columns
# select employee_id rename as emp_id, name, age, salary,tax, columnOne, columntTwo from emp_new_cols

emp_1 = emp_new_cols.withColumnRenamed("employee_id","emp_id") #we can use selectExpr or expr or alias
```

```python
emp_1.show()
```

```python
# Column name with Spaces
# select employee_id as emp_id, name, age, salary, tax, columnOne, columnTwo as 'Column Two' from emp_new_cols
emp_2 = emp_1.withColumnRenamed("columnTwo","Column Two")
```

```python
emp_2.show()
```

```python
# Remove Column

emp_dropped = emp_new_cols.drop("columnTwo","columnOne")
```

```python
emp_dropped.show()
```

```python
# Filter Data
# select employee_id as emp_id, name, age, salary, tax, from emp_dropped where tax>10000

emp_filtered = emp_dropped.select(col("employee_id").alias("emp_id"),"name","salary","tax").where("tax > 10000")
```

```python
emp_filtered.show()
```

```python
# LIMIT data 
# select employee_id as emp_id, name, age, salary, tax from emp_filtered limit 5 

emp_limit = emp_filtered.limit(5).offset(2) # i used offset as well to skip the first two data 
```

```python
emp_limit.show()
```

```python
# we can also use limit as in the view 
emp_limit.show(2)
```

```python
# BONUS TIP
# Add multiple column 
# we can create a dictionary and use withColumns
columns = {
    "tax": col("salary")*0.2,
    "oneNumber": lit(1),
    "twoNumber": lit("two")
}

emp_final = emp.withColumns(columns)
```
