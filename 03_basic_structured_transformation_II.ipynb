```python
# Create DataFrame

emp = spark.createDataFrame(data= emp_data, schema=emp_schema)
```

```python
# Show dataframe (Action)

emp.show()
```

```python
# Print Schema 
emp.printSchema()
```

```python
# Casting Column
# select employee_id, name, age, cast(salary as double) as salary from emp

emp_casted = emp.selectExpr("employee_id", "name", "age", "cast(salary as double) as salary")

emp_casted.printSchema()
```

```python
# Casting Column
# select employee_id, name, age, cast(salary as double) as salary from emp

from pyspark.sql.functions import col

emp_casted = emp.select("employee_id", "name", "age", col("salary").cast("double").alias("salary"))

emp_casted.printSchema()
```
